```python
import numpy as np
import IPython.display as ipd

from datasets import load_dataset
from collections import defaultdict
```

```python
af_dataset = load_dataset("google/fleurs", "af_za") # ~ 4 min
xh_dataset = load_dataset("google/fleurs", "xh_za") # ~ _ min
```

```python
af_train = af_dataset["train"]
af_val = af_dataset["validation"]
af_test = af_dataset["test"]
xh_train = xh_dataset["train"]
xh_val = xh_dataset["validation"]
xh_test = xh_dataset["test"]
```

```python
female = 0
male = 0

lang_groups = defaultdict(int)
for i in range(len(xh_train)):
    lang_group_id = xh_train[i]['lang_group_id'] 
    lang_groups[lang_group_id] += 1

    gender = xh_train[i]['gender'] 
    if gender == 1:
        female += 1
    elif gender == 0:
        male += 1

print(lang_groups)
print(female)
print(male)
```
